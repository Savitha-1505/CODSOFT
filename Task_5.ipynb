```python
# prompt: how to handle the missing values

# Fill missing values with the mean of the column (numerical data)
credit_card_data.fillna(credit_card_data.mean(), inplace=True)

# Alternatively, fill with median
# credit_card_data.fillna(credit_card_data.median(), inplace=True)

# Or fill with a specific value (e.g., 0 for numerical data)
# credit_card_data.fillna(0, inplace=True)

# For categorical data, use mode
for col in credit_card_data.columns:
    if credit_card_data[col].dtype == 'object':
      credit_card_data[col].fillna(credit_card_data[col].mode()[0], inplace=True)

# Verify if missing values are handled
credit_card_data.isnull().sum()
```
